##INTRODUCTION TO EDA

Exploratory Data Analysis(EDA) is a process of examining the data, extracting insights of the data.Originally developed by American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data discovery process today.In statistics, EDA is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods.

##Distplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/audi.project.xlsx')
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.distplot(df['year'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_excel("/content/audi.project.xlsx")
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(df['year'],bins = 10)
plt.title('audi cars')
plt.xlabel('year')
plt.ylabel('price')
plt.show()
```

##Histplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/audi.project.xlsx')
sns.histplot(df['year'],palette='hls')
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
df[['model', 'year', 'price', 'mileage', 'transmission']]
plt.show()
```

##Piechart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_excel('/content/audi.project.xlsx')
df.groupby(['model']).sum().plot( kind='pie', y='price',autopct='%1.0f%%')
plt.show()
```
